```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from scipy.cluster.hierarchy import dendrogram,linkage
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
import pickle
```

```python
df = pd.read_csv('Young people survey.csv')
```

```python
df.head()
```

```python
Target = df['Adrenaline_sports']
```

```python
df = df.drop(['Adrenaline_sports'],axis=1)
```

```python
featuers = df.iloc[:,:-1]
```

```python
pd.options.display.max_columns=None
```

```python
featuers.head()
```

```python
new_df = pd.concat([featuers, Target],axis=1)
```

```python
new_df.head()
```

```python
scaler_df = StandardScaler()
```

```python
featuers_std = scaler_df.fit_transform(featuers)
```

```python
featuers_train, featuers_validation, featuers_test = np.split(featuers_std, [int(.8*len(featuers_std)), int(.9*len(featuers_std))])
Target_train, Target_validation, Target_test = np.split(Target, [int(.8*len(Target)), int(.9*len(Target))])
```

```python
np.savez('survey_train_data',inputs = featuers_train, targets = Target_train)
np.savez('survey_validation_data',inputs = featuers_validation, targets = Target_validation)
np.savez('survey_test_data',inputs = featuers_test, targets = Target_test)
```

```python
pickle.dump(scaler_df, open ('scaler_df.pickle', 'wb'))
```
